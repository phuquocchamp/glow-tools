# FEATURING

````python
features = ["PALSID", "Email", "ETOP Level_Change"]

data = data[features]
data.rename(
    columns = {
        "Email": "email",
        "PALSID": "firstname",
        "ETOP Level_Change": "updated_level"
    },
    inplace=True,
)
````

````python
df = data[data["updated_level"] != "No Change"]
df
````

## BEGINNER

````python
df.loc[df["updated_level"] == "Beginner", "updated_level"] = "ETB-2025"
````

## INTERMEDIATE

````python
df.loc[df["updated_level"] == "Intermediate", 'updated_level'] = "ETI-2025"
````

## ADVANCED

````python
df.loc[df["updated_level"] == "Advanced", "updated_level"] = "ETA-2025"
````

````python
df
````

# MERGEING WITH DATASET

````python
dataset = pd.read_csv("./dataset/standardized_glow_students.csv")

features = ["username", "email", "lastname"]
dataset = dataset[features]
````

````python
final_df = pd.merge(df, dataset, on="email", how="left")
final_df
````

````python
final_df["username"].isnull().sum()
````

# EXPORT TO TARGET FILE

````python
final_df.rename(columns={
    "updated_level": "cohort1"
}, inplace=True)

final_df["role1"] = "student"
final_df.loc[final_df["cohort1"] == "ETB-2025", "course1"] = "ETOP Beginner"
final_df.loc[final_df["cohort1"] == "ETI-2025", "course1"] = "ETOP Intermediate"
final_df.loc[final_df["cohort1"] == "ETA-2025", "course1"] = "ETOP Advanced"
````

````python
final_df
````

````python
final_df.to_csv("./etop_level_updated/final.csv", index=False, encoding='utf-8-sig')
print(f"Export final data succeefuly !")
````
