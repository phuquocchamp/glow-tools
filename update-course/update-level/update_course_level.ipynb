```python
import pandas as pd
```

```python
df = pd.read_csv("./dataset/pronounce/pro_adv.csv")
```

```python
df.rename(columns={
    "Username": "username",
    "First name": "firstname",
    "Surname": "lastname"
}, inplace=True)
```

```python
df.head()
```

```python
# QT18P0085
pattern = r"^[A-Z]{2}[0-9]{2}[A-Z]{1}[0-9]{4}$"

# Kiểm tra PALS_ID có khớp với regex hay không, và loại bỏ các giá trị không khớp
df = df[df["firstname"].str.match(pattern, na=False)]
```

```python
df.head()
```

```python
df = df.drop(columns="Groups")
```

```python
df["course1"] = "Pronunciation - Intermediate"
df["role1"] = "student"
df["cohort1"] = "PROIA-2025"
```
