## IMPORT STANDARD FILE


```python
import os
import pandas as pd
from IPython.display import HTML

monthly_path = "T7"

df_students = pd.read_csv("./data/base-data.csv", encoding="utf-8-sig")

print(df_students.shape)
HTML(df_students.head().to_html())
```

## COURSE_NAME COLUMNS


```python
courses_names = [
  "TOEIC Practice",
  "IELTS Practice",
  "TED Talks - General 1",
  "TED Talks - General 2",
  "TED Talks - General 3",
  "TED Talks AI"
  "TED Talks Career",
  "TED Talks Career Deck",
  "Idioms / Collocation",
  "IELTS Speaking",
  "IELTS Writing",
  "English for Starters",
  "Learn English though Songs",
  "Learn English through Songs - Old",
  "Learn English though Movies",
  "Learn vocabulary with Quizlet",
  "Learn English though Books",
  "Pronunciation - Beginner",
  "Pronunciation - Intermediate",
  "Pronunciation - Advanced",
  "English Placement Assessment",
  "Sentence Diagramming",
  "Grammar",
  "How to use ChatBot AI",
   # "ETOP Orientation for students", 
  "PALS English Self-Learning",
  "Practice for Donor's interview",
  "English Library",
  "Report Non-GLOW English Self-study Time"
]

# Thêm các cột mới với giá trị mặc định là None cho mỗi học sinh
for course in courses_names:
    df_students[course] = 0.0

print("COURSE_NUMBER: ", len(courses_names))
# Hiển thị 20 dòng đầu tiên của DataFrame sau khi thêm cột
print("RECORD_NUMBER: ", len(df_students))
HTML(df_students.head(5).to_html())
```

```python
df_courses = pd.read_csv(
    "./export/2025/" + monthly_path + "/analysic_report.csv", encoding="utf-8-sig"
)

# Lọc các bản ghi chỉ có COURSE_NAME nằm trong danh sách tên các khóa học
df_courses = df_courses[df_courses["COURSE_NAME"].isin(courses_names)]

print("ROW_NUMBER: ", len(df_courses))

HTML(df_courses.head(20).to_html())
```

## COMBINE DATA


```python
# Cập nhật thời lượng cho các khóa học tương ứng
for _, row in df_courses.iterrows():
    pals_id = row["PALS_ID"]
    course_name = row["COURSE_NAME"]
    duration = row["DURATION"] / 60
    if pals_id in df_students["PALS_ID"].values and course_name in df_students.columns:
        df_students.loc[df_students["PALS_ID"] == pals_id, course_name] += float(
            duration
        )


# Hiển thị 20 dòng đầu tiên của DataFrame sau khi cập nhật thời lượng
print("TOTAL_RECORDS: ", len(df_students))
HTML(df_students.head(20).to_html())
```

## EXPORT DATA


```python
import openpyxl

output_file = "./export/2025/" + monthly_path + "/final_report.xlsx"
df_students.to_excel(output_file, index=False, engine="openpyxl")
print("Exported Successfully")
```
