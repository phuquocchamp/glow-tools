## 1. IMPORT LIBRARY

```python
import pandas as pd
import os
import re
from IPython.display import HTML

df = pd.read_excel('./export/valid_data.xlsx')
HTML(df.head(20).to_html())
```

## 2. MERGE DATA

```python
merged_df = (
    df.groupby(["PALS_ID", "FULLNAME"])["DURATION"].sum().reset_index()
)

HTML(merged_df.head(20).to_html())
```

## 3. SAVE FILE

```python
merged_df.to_excel("./export/final/valid_report.xlsx")
print("SAVE FILE SUCCESSFULLY!")
```
